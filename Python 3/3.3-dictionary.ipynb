## Dictionaries

Dictionaries are unordered collections of items, they store data in key-value pairs. key must be unique and immutable (eg: strings,numbers or tuples), while values can be of anytype

```python
## creation of dictionaries

emy_dict ={}
print(type(emy_dict))

emt_dict = dict()
print(type(emt_dict))
```

```python
## keys and values
student ={"name":"Draavith","age":23,"grade":80}

# here keys are name,age and grade and values are Draavith,23 and 80

print(student)
print(type(student))

## errors: key_name should not repeat
```

```python
## accessing dictionaties elements

student ={"name":"Draavith","age":23,"grade":80}
print(student)

print(student['grade'])
print(student["age"])

print(student.get('grade'))
print(student.get("last_name")) # if the key vaule is not present in dictionary then it returns none
print(student.get('mid_name',"invalid input hhehehe")) # if the key value is not present and want to display the default value do this 
```

```python
## modifying Dictionary Elements
## Dictionary are mutable, so you can add, update or delete elements

print(student)
```

```python
student['age'] = 20 ## updating the value
print(student)
student['address']='India' ## adding the new key and value  
print(student)
del student['grade'] ## delete key and value pair
print(student)
```

## dictionary methods


```python
## to get the keys and values and key value pair

keys = student.keys()
print(keys)
values = student.values()
print(values)

items = student.items()
print(items)
```

```python
## shallow copy

student_copy = student
print(student)
print(student_copy)

student['name']="arva"
print(student)
print(student_copy) ## here student copy is also changing its name if i chnage only the student dict. So, to avoid this we use shallow copy


student_copy1 = student.copy() # shallow copy
student['name'] = 'yoyoyoyo'
print(student)
print(student_copy1)

```

```python
## iterating the dictionary

# use loops to itterate

# 1. iterate over the keys
for keys in student.keys():
    print(keys)

for values in student.values():
    print(values)

for key,values in student.items():
    print(f"{keys}:{values}")
```

```python
## Nested dictionaries
students={
    "student1":{"name":"Draavith","age":20,},
    "student2" :{"name":"Ram","age":19},
    "student3": {"name":"Mohan", "age":34}
}
print(students)
```

```python
## access nested dictionaries

print(students["student1"]["age"])
print(students["student2"]["name"])
```

```python
## iterating over nested dictionaries

for student_id,student_info in students.items():
    print(f"{student_id}:{student_info}")
    for keys,values in student_info.items():
        print(f"{keys}:{values}")
```

```python
## dictionary comphrehension

square ={x:x**2 for x in range(5)}
print(square)


## dictionary comphrehension
even_square = {x:x**2 for x in range(10) if x%2==0}
print(even_square)
```

```python
## use dictionary and find the freqency in the list

numbers=[1,1,2,2,3,3,3,4,4,4,4]

frequency={}

for number in numbers:
    if number in frequency:
        frequency[number]+=1
    else:
        frequency[number]=1
print(frequency)
```
