## example 1. manage a to do list

create a to do list to keep track of the tasks

```python
to_do_list=["Buy groceries","clean the room","pay bill"]

## adding to taks
to_do_list.append("schedule meeting")
to_do_list.append("workout")

## removing the compleated taks
to_do_list.remove("clean the room")


## checking if the task in the list 
if "pay bill" in to_do_list:
    print("don't forget to pay the utility bills")

print("to do list remaining")
for task in to_do_list:
    print(f"->{task}")
```
