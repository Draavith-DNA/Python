<h1>Loops</h1>

```python
## for loop: its a loop which itterates through out the range of numbers

for i in range(5):
    print(i)
```

```python
for i in range(5,10):
    print(i)
```

```python
for i in range(1,10,2): 
    print(i)

## here the 1 is the start and 10 is the end and the 2 defines the step ahead
```

```python
## for in strings

str ="Draavith"

for i in str:
    print(i)
```

```python
## while: this is the loop that continues to execute as long as the condition is true

count =0

while count<6:
    print(count)
    count += 1
```

## loop control statments

```python
## 1. break: this statment exists the loops prematurely

for i in range (10):
    if i == 6:
        break
    print(i)
```

```python
## continue: skips the current iteration and continues with the next

for i in range(10):
    if i%2==0:
        continue
    print(i)
```

```python
## pass: this is the null operation it basically does nothing

for i in range(10):
    if i%2==0:
        pass
    print(i)
```

```python
## nesteredd loop: Ther loop inside the loop is calles as nestered loop

for i in range (3):
    for j in range(2):
        print(f"i:{i} and j:{j}")
```

## examples:


```python
## 1. calculate the sum of 1st n natural number using while and for loop

# while:
num =int(input("Enter the number"))
sum = 0
while num > 0:
    sum += num
    num -= 1
print(sum)
```

```python
# for:
num =int(input("Enter the number"))
sum = 0
for i in range (num+1):
    sum += i
print(sum)
```

```python
## 2. Display the prime numbers between 1 and 100 

for i in range (1,100+1):
    if i>1:
        for j in range (2, i):
            if i % j == 0:
                break
        else:
            print(i)
```
