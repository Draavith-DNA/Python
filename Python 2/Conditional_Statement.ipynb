```python
## elif (else-if): the elif statment allows you to check multiple conditions.

age= 23

if age < 10:
    print("age is less than 10")
elif age > 10 and age <=25:
    print("age is greater then 10 and less than or equal to 25")
else:
    print("age is greater than 50")
```

```python
## nested conditional statments: placing one or more if, elif, or else statments inside another if, elif or else statment is called as the nested conditional statments

## problem: given numbver is even , odd or negative


num=int(input("Enter the number"))

if num>0:
    print("the number is positive")
    if num%2==0:
        print("the number is even")
    else:
        print("the number is odd")
else:
    print("the number is negative or zero")
```

```python
## Problem: determine if the year is a leap year using the nested conditioanl statment

year=int(input("enter the year"))

if year % 4 == 0:
    if year % 100 == 0:
        if year %400 == 0:
            print("it is the leap year")
        else:
            print("not a leap year")
    else:
        print("not a leap year")       
else:
    print("the year is not leap year")
```
