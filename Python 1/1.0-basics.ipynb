Indentation in Python means adding spaces or tabs at the beginning of a line to define the block of code.
Python uses indentation to decide which statements belong to the same block (like inside loops, functions, conditions, classes, etc.).

Unlike other languages that use { } (like C++, Java, JavaScript),
Python uses indentation to determine the structure and flow of the program.

```python
## indentation example 

age = 19
if age>10:
    print(age)

print ("age is below 10")


## In cpp and java we use { } insted of : after the condition
```
